```python
#Définition de la carte et de ses attributs de base
import folium
from folium.plugins import MiniMap, MousePosition, MeasureControl, TagFilterButton


# Carte principale
m = folium.Map(
    max_bounds=True,
    location=[48.862945, 2.349594],
    zoom_start=12,
    tiles="cartodbpositron",
    control_scale=True
)

restaurants_group = folium.FeatureGroup(name="Restaurants", show=True)

# Fonction générique pour ajouter les points
def add_markers(df, type_name):
    if df is None:  # Si pas de DataFrame pour ce type
        return
    color = type_color_mapping.get(type_name, "gray")


    for _, row in df.iterrows():
        # Use 'Type' column for all dataframes
        type_column = 'Type'

        # Création du contenu HTML pour le popup
        popup_html = f"""
        <div style="font-size:13px; line-height:1.4;">
            <p><b>Nom :</b> {row['Nom']}<br></p>
            <p><b>Métro :</b> {row['Métro']}<br></p>
            <p><b>Adresse :</b> {row['Adresse']}</p>
        """
        # Add 'Notes' to popup if the column exists
        if 'Notes' in row:
            popup_html += f"<p><b>Note :</b> {row['Notes']}</p>"

        popup_html += "</div>"


        # Tooltip simple
        tooltip_text = f"{row['Nom']} ({row[type_column]})"

        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            popup=folium.Popup(popup_html, max_width=250),
            tooltip=tooltip_text,
            tags=[row[type_column]],
            fill_color=color,
            color="none",
            radius=4,
            fill_opacity=1
        ).add_to(restaurants_group)

# Ajout des marqueurs pour chaque DataFrame
for type_name, df in dataframes.items():
    add_markers(df, type_name)

restaurants_group.add_to(m)

Bars = folium.GeoJson(
    gdg,
    marker=folium.CircleMarker(radius=4, fill_color="grey", fill_opacity=1, color="none", weight=1), # Remove tags=[data]
    name="Bars",
    tooltip=folium.GeoJsonTooltip(
        fields=["Type","Nom", "Arr.","Adresse", "Métro"],
        aliases=["Type","Nom", "Arrondissement","Addresse", "Métro"],
    ),
    popup=folium.GeoJsonPopup(fields=["Nom", "Arr.", "Adresse", "Métro", "Note"])
).add_to(m)

# Plugins
MiniMap(toggle_display=True).add_to(m)
MousePosition().add_to(m)
m.add_child(MeasureControl())
folium.plugins.LocateControl().add_to(m)
folium.plugins.Fullscreen(
    position="topright", title="Expand me", title_cancel="Exit me",
    force_separate_button=True
).add_to(m)

# Couches
folium.TileLayer("CartoDB Positron").add_to(m)
folium.TileLayer("OpenStreetMap").add_to(m)
folium.TileLayer("CartoDB Voyager", show=True).add_to(m)
folium.LayerControl().add_to(m)

# Filtres par tags
unique_tags = gdf['Type'].unique().tolist()
TagFilterButton(unique_tags).add_to(m)
m
```
